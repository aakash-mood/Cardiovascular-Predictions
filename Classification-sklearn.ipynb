```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
breast_cancer
```

```python
breast_df = pd.DataFrame(breast_cancer["data"] ,columns=breast_cancer["feature_names"])

breast_df["target"] = breast_cancer["target"]
breast_df.head()
```

```python
### Import the model
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

## Np seed
np.random.seed(42)

## Create X and Y and get your data ready
X = breast_df.drop("target", axis=1)
Y = breast_df["target"]

## Split the trainning and test data separetly

X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                   Y,
                                                   test_size=0.2)

## Instantiate the classifier and fit the training data set

clf = LinearSVC(max_iter=10000)
clf.fit(X_train,Y_train)

## Evaluate the performance of the model
clf.score(X_test, Y_test)
```

```python
### Importing another classifer using the sklearn-map-cheat-sheet

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

## Np seed
np.random.seed(42)

## Create X and Y and get your data ready
X = breast_df.drop("target", axis=1)
Y = breast_df["target"]

## Split the trainning and test data separetly

X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                   Y,
                                                   test_size=0.2)

## Instantiate the classifier and fit the training data set

clf = RandomForestClassifier()
clf.fit(X_train,Y_train)

## Evaluate the performance of the model
clf.score(X_test, Y_test)
```

#### In case of Breast cancer classification estimators, one can see that the performance of the RandomForestClassifier by passes the 
#### LinearSVC
